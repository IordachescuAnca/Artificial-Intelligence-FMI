```python
os.mkdir('../submission_files/Task2')
folder_test_output = '../submission_files/Task2'
folder_tracking = 'yolov5/runs/detect/yolo_bowl_results/labels'
for fname in os.listdir(base_folder):
    complete_filename = base_folder + '/' + fname
    if fname[-4:] == '.txt':
        #for each video we open the .txt file to get the number of all frames
        print(complete_filename)
        f = open(complete_filename, 'r')
        text = f.readlines()
        f.close()
        first_line = text[0]
        frames = []
        #after that, we select from the label tracking folder the .txt
        #that contains the bounding box of each frame for each video
        for fname_label in os.listdir(folder_tracking):
            if fname_label[:2] == fname[:2]:
                #print(fname_label[3:-4])
                frames.append(int(fname_label[3:-4]))
        frames.sort()
        #get the number of frames in which we detected a bowling ball
        last_frame = frames[-1]

        output_write = first_line
        #iterate through all frames until maximum number of frame
        for i in range(1, last_frame+1):
            #if there is a frame at position i in which the tracker didnt
            #get any bounding box, i am going to select the bounding box that appears in the first frame after frame i
            indx_i = -1
            if i not in frames:
                indx_i = i
                for fr in frames:
                    if fr > i:
                        indx_i = fr
                        break
                #print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!flag: {}, {}".format(i, indx_i))
            else:
                indx_i = i

            #get the coordinates in yolov5 format and
            #convert them to x_min, y_min, x_max, y_max
            file_label = folder_tracking + '/' + fname[:2] + '_' + str(indx_i) + '.txt'
            f_l = open(file_label, 'r')
            text = f_l.readlines()[-1].split()
            f_l.close()
            x_c = float(text[1])
            y_c = float(text[2])
            h = float(text[3])
            w = float(text[4])

            new_x_min = int(x_c *1280 - h*1280/2)
            new_y_min = int(y_c*720 - w*720/2)
            new_x_max = int(x_c *1280 + h*1280/2)
            new_y_max = int(y_c*720 + w*720/2)
            out = str(i-1) + ' ' + str(new_x_min) + ' ' + str(new_y_min) + ' ' + str(new_x_max) + ' ' + str(new_y_max)
            #write the bounding box at the required position in a file
            output_write = output_write + out + '\n'

        new_file = open(folder_test_output+'/'+fname[:2] +'_predicted.txt', 'w')
        new_file.write(output_write)
        new_file.close()
```
