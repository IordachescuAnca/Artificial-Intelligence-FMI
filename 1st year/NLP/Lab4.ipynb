1. Write a function that returns all synonyms for a word in WordNet, given the word string and its language. Return as a list of strings, irrespective of the sense of each word (if word has multiple senses, all synonyms are stil grouped together)

```python
from nltk import download
download('wordnet')
download('omw')

from nltk.corpus import wordnet as wn
```

```python
def all_synonyms(word, lang='eng'):
  synonyms_list = set()
  for synset in wn.synsets(word, lang=lang):
    synonyms = synset.lemma_names(lang)
    synonyms_list.update(synonyms)
  return list(synonyms_list)
  
syn_eng = all_synonyms(word="small", lang='eng')
syn_span = all_synonyms(word="chico", lang='spa')
print("Synonym in English for small: {}".format(syn_eng))
print("Synonym in Spanish for chico: {}".format(syn_span))
```

2. Write a function that retrieves the "translation" of an English word to a different language (one supported by OMW), according to WordNet. (could return multiple words)

```python
def translation(word, lang='eng'):
  translation_list = set()
  for synset in wn.synsets(word):
    lemmas = synset.lemmas(lang=lang)
    words = [l.name() for l in lemmas]
    translation_list.update(words)
  return list(translation_list)

translation('nail', lang='ita')
```

3. Write a function that retrieves the "sibling" terms of a noun in WordNet. A sister term is one that shares a hypernym. (e.g. "oak tree" and "fir tree").

```python
def sister_terms(word):
  synsets = wn.synsets(word, pos='n')
  hypernyms = set()
  #compute the hypernyms of the word
  for synset in synsets:
    hypernyms.update(synset.hypernyms())

  #compute the hyponyms of the hypernyms to get sister words
  hyponyms = set()
  for synset in hypernyms:
    hyponyms.update(synset.hyponyms())
  
  #eliminate the synsets of the orginal word
  sister_words_synsets = list(set(hyponyms) - set(list(synsets)))
  sister_words = set()
  for synset in sister_words_synsets:
    sisters_words = synset.lemma_names()
    sister_words.update(sisters_words)
  
  return list(sister_words)

sister_words = sister_terms('oak')
print(sister_words)
#verify if fir tree and oak tree are sister terms
print('fir' in sister_words)
```

6. Write a function to compute the similarity between two words, using one of the similarity metrics defined in WordNet. If a word as more than one sense, use a method to aggregate the per-synset similarities (e.g. average the distance across all pairs).


```python
def similarity(word1, word2):
  word1_synsets = wn.synsets(word1)
  word2_synsets = wn.synsets(word2)

  score = 0
  #number of pairs of synsets
  pairs = 0
  for synset1 in word1_synsets:
    for synset2 in word2_synsets:
      #we can apply lch_similarity if the words have the same pos
      if synset2.pos() == synset1.pos():
        if synset1.lch_similarity(synset2) is not None:
          #check if it can be calculated
          score += synset1.lch_similarity(synset2)
          pairs += 1
  
  if pairs == 0:
    #didnt get any score and return score 0
    return 0
  return score / pairs

print(similarity('roof', 'home'))
```

7. Compare with word2vec similarity. (Load an embeddings model to be able to retrieve similarities between words). Then take the vocabulary from Ex 5, or a custom vocabulary of your choice (e.g. the SimLex set of word pairs) and compute similarities between all word pairs using the function from Ex 6. Then compute the same similarities with word2vec, and finally compute the Spearman correlation between them.

```python
import itertools
import gensim
import gensim.downloader
from scipy.stats import spearmanr

text = ' '.join(brown.words()[:120])
text = text.lower()
words = word_tokenize(text)
words = list(set(words))
#create all the combinations of words having length = 2
pairs = list(itertools.combinations(words, 2))
embeddings_twitter = gensim.downloader.load('glove-twitter-25')

lch_sim = []
emb_sim = []

for pair in pairs:
  if pair[0] in embeddings_twitter and pair[1] in embeddings_twitter:
    #we calculate the score only for words that has embeddings in order to get two vectors that have the same size (lch_sim and emb_sim)
    score_emb = embeddings_twitter.similarity(pair[0], pair[1])
    emb_sim.append(score_emb)

    score_lch = similarity(pair[0], pair[1])
    lch_sim.append(score_lch)

coef, p = spearmanr(lch_sim, emb_sim)
print("Spearman Correlation: {}".format(coef))
```

8. Using the same text from exercise 5, select all nouns (according to WordNet, choose words which are part of any noun Synsets) and find their hypernyms. Reconstruct the text replacing each noun with one of its hypernyms. Is it coherent? Pinpoint some examples where the replacement did not make sense (in a text cell or a comment)

```python
text = ' '.join(brown.words()[:120])
text = text.lower()
words = word_tokenize(text)
new_text = ""
for word in words:
  synsets = wn.synsets(word)
  hypernyms_name = []
  has_any_noun_sunsets = False
  for synset in synsets:
    if synset.pos() == 'n':
      #check if the word has any noun synset
      has_any_noun_sunsets = True
  
  if has_any_noun_sunsets == False:
    #the word remains the same
    new_text = new_text + word + ' '
  else:
    #get all the hypernyms of that word
    hypernyms_name = []
    for synset in synsets:
      hypernyms = synset.hypernyms()
      for hypernym in hypernyms:
        hypernym_n = hypernym.lemma_names()
        hypernyms_name = hypernyms_name + hypernym_n
    
    if len(hypernyms_name) == 0:
      #the word doesnt have any hypernyms and it remains the same
      new_text = new_text + word + ' '
    else:
      #i choose a hypernym from all the hypernyms i got
      new_text = new_text + hypernyms_name[0] + ' '

print("Original text: {}".format(text))
print("New text: {}". format(new_text))

```

Original: the jury further said in term-end presentments that the city executive committee

Examples of replacements which don't work: the body further said linear_unit term-end due_process that the municipality administrator administrative_unit
